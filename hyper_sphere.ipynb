```python
import numpy as np


import numpy as np


def generate_points_on_hypersphere(n_points, n_dimensions):
    points = np.empty((n_points, n_dimensions))

    for i in range(n_points):
        # Generate random points on the unit sphere
        x = np.random.normal(size=n_dimensions)
        x /= np.linalg.norm(x)

        # Scale points to the hypersphere's radius
        radius = 1.0
        point = radius * x

        points[i] = point

    return points
```

```python
from matplotlib import pyplot as plt

data = generate_points_on_hypersphere(100, 5)


# Extract x, y, and z coordinates from the points
x = [point[0] for point in data]
y = [point[1] for point in data]
z = [point[2] for point in data]

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Plot the points
ax.scatter(x, y, z, c="b", marker=".")

# Set plot labels and title
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Points on a 3D Hypersphere")

# Show the plot
plt.show()
```
